```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('BankChurners.csv')
```

```python
df.index = df['CLIENTNUM']
```

```python
df_c = df[df.columns[:-2]]
```

```python
df_c.describe()
```

```python
df_c["Months_on_book"].value_counts().plot(kind='bar')
plt.xlabel('Count', labelpad=14)
plt.ylabel('Target Variable', labelpad=14)
plt.title('Target Variable', pad=14)
```

```python
df_c["Customer_Age"].value_counts().plot(kind='bar')
plt.xlabel('Count', labelpad=14)
plt.ylabel('Target Variable', labelpad=14)
plt.title('Target Variable', pad=14)
```

```python
fig = plt.figure(figsize=(12, 6))
plt.scatter(df_c["Customer_Age"],df_c["Avg_Utilization_Ratio"], s=2)
plt.title("Size")
plt.xlabel("Customer_Age")
plt.ylabel("Avg_Utilization_Ratio")
```

```python
df_train = df_c.sample(frac=0.8, random_state=0)
df_test = df_c.drop(df_train.index)
print(len(df_train))
print(len(df_test))
```

```python
df_train["Income_Category"] = df_train["Income_Category"].replace({"<=50K":0, ">50K":1})
sns.boxenplot(x="Income_Category", y="Avg_Utilization_Ratio", data=df_train)
plt.ylim(0,1)
```
